```python
import pandas as pd
import numpy as np
import re
import geopandas as gpd
import os
import folium
import json
```

```python
gwangjin = gpd.read_file("/석사과정/3학기/광진구_집계구_2023.shp", encoding = "CP949")
gwangjin
```

### 사용 가능한 색상
‘red’, ‘blue’, ‘green’, ‘purple’, ‘orange’, ‘darkred’, ’lightred’, ‘beige’, ‘darkblue’, ‘darkgreen’, ‘cadetblue’, ‘darkpurple’, ‘white’, ‘pink’, ‘lightblue’, ‘lightgreen’, ‘gray’, ‘black’, ‘lightgray’

### 선 스타일
- 실선: "0, 0" / "solid"
- 대시선: "5, 5" / "dashed"
- 점선: "2, 2" / "dotted"
- 대시 점선: "5, 3, 2, 3" / "dashdot"

```python
gu = gwangjin[gwangjin['ADM_NM']=='구의1동']
center = [gu['geometry'].centroid.y.mean(), gu['geometry'].centroid.x.mean()]
map_geo = folium.Map(location=center, zoom_start=20)

def style_function(feature):
    return {
        'fillColor': 'white',  
        'color': 'orange',     
        'weight': 4,            
        #'dashArray': 'dotted'     
    }

folium.GeoJson(gu, style_function=style_function).add_to(map_geo)
map_geo
```

```python
gu_mart = pd.read_csv("구의1동_마트.csv", encoding = 'euc-kr')
gu_mart
```

```python
for i in range(gu_mart.shape[0]):
    folium.Marker(
        [gu_mart['위도'][i], gu_mart['경도'][i]], 
        icon=folium.Icon(color='red',icon="fa fa-shopping-cart",prefix='fa-solid')
    ).add_to(map_geo)
    
    folium.Circle(
    [gu_mart['위도'][i], gu_mart['경도'][i]],
    radius=100,
    color='red',
    fill_color='red'
    ).add_to(map_geo)
    
map_geo
```
