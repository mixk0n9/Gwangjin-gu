```python
def get_hashtags(driver):
    # 해시태그 요소들 찾기
    hashtag_elements = driver.find_elements(By.CLASS_NAME, "x16tdsg8")

    # 해시태그 텍스트 추출
    hashtags = [element.text for element in hashtag_elements if '#' in element.text]

    return hashtags

hashtags = get_hashtags(driver)
print("해시태그:", hashtags)
```

```python
def get_content(driver):
    # 본문 텍스트 가져오기
    content_element = driver.find_element(By.CLASS_NAME, "_aad7")
    content_text = content_element.text

    # datetime 요소 가져오기
    datetime_element = driver.find_element(By.CLASS_NAME, "_a9ze")

    # datetime 속성 값 가져오기
    datetime_value = datetime_element.get_attribute("datetime")

    # 날짜 부분 추출하기
    date = datetime_value[:10]

    # 해시태그 가져오기
    hashtags = get_hashtags(driver)

    return content_text, date, hashtags
```

```python
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
import time 


data = []


while True:

    content_text, date, hashtags = get_content(driver)
    print("본문 내용:", content_text)
    print("작성 날짜:", date)
    print("해시태그:", hashtags)
    
    new_post = {'Content': content_text, 'Date': date, 'Hashtags': hashtags}
    data.append(new_post)
    
    # 다음 게시물로 이동
    try:
        nextBtn = driver.find_element(By.CSS_SELECTOR, "body > div.x1n2onr6.xzkaem6 > div.x9f619.x1n2onr6.x1ja2u2z > div > div.x1uvtmcs.x4k7w5x.x1h91t0o.x1beo9mf.xaigb6o.x12ejxvf.x3igimt.xarpa2k.xedcshv.x1lytzrv.x1t2pt76.x7ja8zs.x1n2onr6.x1qrby5j.x1jfb8zj > div > div > div > div > div:nth-child(1) > div > div > div._aaqg._aaqh > button")
        nextBtn.click()
        
        time.sleep(8)
    except NoSuchElementException:
        print("마지막 게시물입니다.")
        break


df = pd.concat([pd.DataFrame([post]) for post in data], ignore_index=True)

print(df)
```
